#### A House can be called a luxury hourse iff

1. Condition of the House if 'Good' or 'Excellent
2. overall grade >= 8
3. Waterfront_View is present

```python
Housing['luxury_home'] = 0

for i in range(len(Housing)):
  count = 0
  if Housing['Waterfront View'][i] == 'Yes':
    count = count+1
  if Housing['Condition of the House'][i] in ['Good','Excellent']:
    count = count+1    
  if Housing['Overall Grade'][i] >= 8:
    count = count+1
  if count >= 2:
    Housing['luxury_home'][i] = 'Yes'
  else:
    Housing['luxury_home'][i] = 'No'
    
Housing['luxury_home'].unique()
```

```python
#apply method

def luxury_home(row):
    count=0
    if row[0]=='Yes':
        count=count+1
    if row[1] in ['Good','Excellent']:
        count=count+1
    if row[2] >= 8:
        count=count+1
        
    if count >=2:
        return 'Yes'
    else:
        return 'No'

Housing['luxury_home']=Housing[['Waterfront View','Condition of the House','Overall Grade']].apply(luxury_home, axis=1)
Housing['luxury_home'].unique()
```

### Treating Outliers

1. Deletion
2. Imputing
3. Data Transformation

```python
#imputing
def limit_imputer(value):
    if value > upper_limit:
        return upper_limit
    if value < lower_limit:
        return lower_limit
    else:
        return value
Housing['Sale Price']=Housing['Sale Price'].apply(limit_imputer)
Housing['Sale Price'].describe()
```
